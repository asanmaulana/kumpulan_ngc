```python
import pandas as pd 
from selenium import webdriver 
from bs4 import BeautifulSoup
from time import sleep
```

```python
#instance objek untuk webdriver
driver = webdriver.Chrome()

#link yang akan di akses
url = "https://coinmarketcap.com/"

driver.implicitly_wait(10)
#ini untuk ambil link sementara bro
driver.get(url)


for scroll in range(30):
        driver.execute_script("window.scrollBy(0,350)") #sekali scroll 250 pixel
        sleep(1)


#ini merupakan kerangka html yang masih sebaris
html = driver.page_source 

soup = BeautifulSoup(html, "html.parser")
#prettify disini menunjukkan 700 karakter pertama bukan karakter yang akan kita ambil
#yang kita ambil tetap semuanya ges ya
print(soup.prettify()[:700])
```

```python
import pandas as pd

#inisiasi dataframe kosong untuk menampung hasil scraping
data = pd.DataFrame()

#kolom baru (title) yang menangkap looping scrape coin tersebut

data_nama = [ title.get_text() for title in soup.find_all('div',{"class":"sc-4c05d6ef-0 bLqliP"}) ]
data_harga = [ title.get_text() for title in soup.find_all('div', {"class": "sc-b3fc6b7-0 dzgUIj"})]
data_marketcap = [ title.get_text() for title in soup.find_all('span',{"class":"sc-11478e5d-1 jfwGHx"}) ]
data_volume = [ title.get_text() for title in soup.find_all('p',{"class":"sc-71024e3e-0 bbHOdE font_weight_500"}) ]
data_coinsupply = [ title.get_text() for title in soup.find_all('div',{"class":"sc-4c05d6ef-0 sc-552ef8b1-0 dlQYLv eOHbrv"}) ]

#menentukan variabel mana yang mempunyai data paling panjang dan kemudian data paling panjang itu jadi acuan 
# data acuan itu akan kita gunakan jika daata lain ga ada akan diisi missing value saja

max_len = max(len(data_nama), len(data_harga), len(data_marketcap), len(data_volume), len(data_coinsupply))

# Standardize lists by filling missing values
fill_value = None  # This could be set to a default string like 'Not Available'
data_nama += [fill_value] * (max_len - len(data_nama))
data_harga += [fill_value] * (max_len - len(data_harga))
data_marketcap += [fill_value] * (max_len - len(data_marketcap))
data_volume += [fill_value] * (max_len - len(data_volume))
data_coinsupply += [fill_value] * (max_len - len(data_coinsupply))


# kita coba untuk standarisasi prosesnya 
# sekarang masukkan data itu ke dataframe

data['nama_coin'] = data_nama
data['harga_coin'] = data_harga
data['market_cap'] = data_marketcap
data['volume'] =data_volume
data['coin_supply'] = data_coinsupply 

data
```

```python
import pandas as pd 
from selenium import webdriver 
from bs4 import BeautifulSoup
from time import sleep

namakoin =[]
hargakoin = []
marketcapital = []
volume = []
koinsupply = []

driver = webdriver.Chrome()

for i in range(1,11) :
    url="https://coinmarketcap.com/?page={}".format(i)

    sleep(2)

    
    driver.get(url)

    for scroll in range(30):
        driver.execute_script("window.scrollBy(0,350)") #sekali scroll 350 pixel
        sleep(1)


    html = driver.page_source
    soup = BeautifulSoup(html, "html.parser")
    sleep(3)

    #ambil data masing masing elemen
    namakoin += [ title.get_text() for title in soup.find_all('div',{"class":"sc-4c05d6ef-0 bLqliP"}) ]
    hargakoin += [ title.get_text() for title in soup.find_all('div', {"class": "sc-b3fc6b7-0 dzgUIj"})]
    marketcapital += [ title.get_text() for title in soup.find_all('span',{"class":"sc-11478e5d-1 jfwGHx"}) ]
    volume += [ title.get_text() for title in soup.find_all('p',{"class":"sc-71024e3e-0 bbHOdE font_weight_500"}) ]
    koinsupply += [ title.get_text() for title in soup.find_all('div',{"class":"sc-4c05d6ef-0 sc-552ef8b1-0 dlQYLv eOHbrv"}) ]




```
